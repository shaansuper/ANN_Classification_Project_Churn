```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}

## We need to pass this data , but first we need to transform and scale data. SO now
##lets load the gender encoders and geography one hot encoding
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
```

```python
my_model = load_model('model.h5')

## Reading the label encoder and one hot encoder as well as scaler


with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('one_hot_encoder_geo.pkl','rb') as file:
    one_hot_encoder_geo=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)



```

```python
# Transform the input data to Dataframe

data=pd.DataFrame([input_data])
```

```python
data
```

```python
data['Gender']=label_encoder_gender.transform(data['Gender'])
```

```python
data
```

```python
array_geography=one_hot_encoder_geo.transform(data[['Geography']]).toarray()

geo_encoded_df=pd.DataFrame(array_geography,columns=one_hot_encoder_geo.get_feature_names_out())
```

```python
geo_encoded_df
```

```python
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
data
```

```python
input_scaled=scaler.transform(data)
input_scaled
```

```python
prediction=my_model.predict(input_scaled)
```

```python
prediction
```

```python
prediction_probability=prediction[0][0]
```

```python
if prediction_probability>0.5:
    print("Will churn")
else:
    print("Won't Churn")
```
